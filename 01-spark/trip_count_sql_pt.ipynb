```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("trip_count_sql").getOrCreate()
```

```python
directory = "/Users/gimhyeonjeong/data-engineering/01-spark/data"
filename = "fhvhv_tripdata_2020-03.csv"
```

```python
data = spark.read.csv(f"file:///{directory}/{filename}", inferSchema = True, header= True)
```

```python
data.show(5)
```

```python
data.createOrReplaceTempView("mobility_data")
```

```python
spark.sql("select *from mobility_data limit 5").show()
```

```python
spark.sql("select pickup_datetime from mobility_data").show()
```

```python
spark.sql("select split(pickup_datetime, ' ')[0] as pickup_data from mobility_data").show()
```

```python
spark.sql("select pickup_data, count(*) as trips from (select split(pickup_datetime, ' ')[0] as pickup_data from mobility_data) group by pickup_data").show()
```
